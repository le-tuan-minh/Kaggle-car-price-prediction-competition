```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

# Giả định rằng DataFrame 'df' của bạn đã được tải và các bước tiền xử lý trước đó đã chạy.
# Nếu bạn chạy toàn bộ code này, bạn sẽ cần đảm bảo df tồn tại và có dữ liệu.

# --- Bắt đầu tiền xử lý các biến (đã bao gồm các bước trước đó) ---

print("--- Bắt đầu tiền xử lý dữ liệu ---")

# =======================================================
# 1. Biến 'name'
# =======================================================
print("\nĐang xử lý biến 'name'...")
if 'name' in df.columns:
    df['name'] = df['name'].astype(str).str.split(' ').str[0]
    print("Đã giữ lại từ đầu tiên trong cột 'name'.")
else:
    print("Cột 'name' không tồn tại. Bỏ qua.")


# =======================================================
# 2. Biến 'mileage'
# =======================================================
print("\nĐang xử lý biến 'mileage' (QUY ĐỔI CHÍNH XÁC về kmpl dựa trên 'fuel')...")

if 'mileage' in df.columns and 'fuel' in df.columns:
    # Tách giá trị số và đơn vị trong cột mileage
    extracted_data = df['mileage'].astype(str).str.extract(r'(\d+\.?\d*)\s*([a-zA-Z\/]*)')
    df['mileage_numeric'] = pd.to_numeric(extracted_data[0], errors='coerce')
    df['mileage_unit'] = extracted_data[1].str.strip().replace('', np.nan).str.lower()

    # Khởi tạo cột quy đổi
    df['mileage_kmpl_equivalent'] = np.nan

    # Hệ số quy đổi từ km/kg sang kmpl
    CNG_TO_LITERS = 1.35  # 1 kg CNG ≈ 1.35 liters
    LPG_TO_LITERS = 1.84  # 1 kg LPG ≈ 1.84 liters

    # Trường hợp mileage đã là kmpl
    kmpl_mask = df['mileage_unit'] == 'kmpl'
    df.loc[kmpl_mask, 'mileage_kmpl_equivalent'] = df.loc[kmpl_mask, 'mileage_numeric']

    # Trường hợp km/kg + CNG
    cng_mask = (df['mileage_unit'] == 'km/kg') & (df['fuel'].str.lower() == 'cng')
    df.loc[cng_mask, 'mileage_kmpl_equivalent'] = df.loc[cng_mask, 'mileage_numeric'] / CNG_TO_LITERS

    # Trường hợp km/kg + LPG
    lpg_mask = (df['mileage_unit'] == 'km/kg') & (df['fuel'].str.lower() == 'lpg')
    df.loc[lpg_mask, 'mileage_kmpl_equivalent'] = df.loc[lpg_mask, 'mileage_numeric'] / LPG_TO_LITERS

    # Trường hợp lỗi: Petrol hoặc Diesel mà lại là km/kg
    invalid_mask = (df['mileage_unit'] == 'km/kg') & (df['fuel'].str.lower().isin(['petrol', 'diesel']))
    if invalid_mask.any():
        print(f"[!] Cảnh báo: Có {invalid_mask.sum()} dòng có 'km/kg' nhưng nhiên liệu là Petrol hoặc Diesel. Gán NaN do lỗi dữ liệu.")
        df.loc[invalid_mask, 'mileage_kmpl_equivalent'] = np.nan

    # Các trường hợp còn lại không rõ loại nhiên liệu (vẫn là km/kg)
    unknown_mask = (df['mileage_unit'] == 'km/kg') & (~df['fuel'].str.lower().isin(['cng', 'lpg', 'petrol', 'diesel']))
    if unknown_mask.any():
        print(f"[!] Cảnh báo: Có {unknown_mask.sum()} dòng có đơn vị km/kg nhưng không xác định được loại nhiên liệu. Không quy đổi.")

    print("✅ Đã quy đổi chính xác cột 'mileage' về đơn vị kmpl theo loại nhiên liệu.")
else:
    print("❌ Cột 'mileage' hoặc 'fuel' không tồn tại. Bỏ qua xử lý.")



# =======================================================
# 3. Biến 'engine'
# =======================================================
print("\nĐang xử lý biến 'engine'...")
if 'engine' in df.columns:
    df['engine'] = df['engine'].astype(str).str.replace('[^0-9.]', '', regex=True)
    df['engine'] = pd.to_numeric(df['engine'], errors='coerce')
    print("Đã xử lý cột 'engine', loại bỏ đơn vị và chuyển về kiểu số.")
else:
    print("Cột 'engine' không tồn tại. Bỏ qua.")

# =======================================================
# 4. Biến 'max_power'
# =======================================================
print("\nĐang xử lý biến 'max_power'...")
if 'max_power' in df.columns:
    df['max_power'] = df['max_power'].astype(str).str.lower()
    df['max_power'] = df['max_power'].replace(r'^\s*bhp\s*$', np.nan, regex=True)
    df['max_power'] = df['max_power'].str.replace('[^0-9.]', '', regex=True)
    df['max_power'] = pd.to_numeric(df['max_power'], errors='coerce')
    print("Đã xử lý cột 'max_power', loại bỏ 'bhp', chuyển 'bhp' không có số thành NaN và chuyển về kiểu số.")
else:
    print("Cột 'max_power' không tồn tại. Bỏ qua.")

# =======================================================
# 5. Biến 'torque'
# =======================================================
print("\nĐang xử lý biến 'torque'...")

if 'torque' in df.columns:
    df['torque_Nm'] = np.nan
    df['torque_unit_extracted'] = np.nan
    processed = df['torque'].astype(str).str.lower().str.strip()

    for idx, val in processed.items():
        # Bước 1: Tách số đầu tiên
        match = re.search(r'(\d+\.?\d*)', val)
        unit = None

        if match:
            number = float(match.group(1))

            # Bước 2: Cố gắng bắt đơn vị sau số
            unit_match = re.search(rf"{re.escape(match.group(1))}\s*(@|at)?\s*\d*[,]?\d*\s*\(?\s*(kgm|nm|kg\.m|n\.m)", val)
            if unit_match:
                raw_unit = unit_match.group(2).replace('.', '')
                unit = raw_unit
            else:
                # Nếu không có đơn vị ngay sau số, tìm đơn vị ở bất kỳ đâu
                fallback_unit = re.search(r'(kgm|nm|kg\.m|n\.m)', val)
                if fallback_unit:
                    unit = fallback_unit.group(1).replace('.', '')

            # Bước 3: Xử lý đơn vị và quy đổi
            if unit == 'kgm':
                df.at[idx, 'torque_Nm'] = number * 9.80665
                df.at[idx, 'torque_unit_extracted'] = 'kgm'
            elif unit == 'nm':
                df.at[idx, 'torque_Nm'] = number
                df.at[idx, 'torque_unit_extracted'] = 'Nm'
            else:
                # Không rõ đơn vị: giữ nguyên số, nhưng đơn vị = NaN
                df.at[idx, 'torque_Nm'] = number
                df.at[idx, 'torque_unit_extracted'] = np.nan
        else:
            # Không có số nào: bỏ qua
            df.at[idx, 'torque_Nm'] = np.nan
            df.at[idx, 'torque_unit_extracted'] = np.nan

    print("✅ Đã xử lý xong 'torque', kể cả các dạng có đơn vị nằm trong ngoặc.")
else:
    print("❌ Cột 'torque' không tồn tại.")
# Gán 'Nm' cho những giá trị lớn nhưng chưa rõ đơn vị
df.loc[(df['torque_Nm'] > 100) & (df['torque_unit_extracted'].isna()), 'torque_unit_extracted'] = 'Nm'
print("\n--- Hoàn tất tiền xử lý ---")

# =======================================================
# 6. Làm tròn các giá trị numeric đến 3 chữ số thập phân
# =======================================================
print("\nĐang làm tròn các giá trị số trong DataFrame đến 3 chữ số thập phân...")
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    df[col] = df[col].round(3)
print("Đã làm tròn các cột số.")
print(df[numeric_cols].head())

# =======================================================
# 7. Visualize lượng missing data theo từng biến
# =======================================================
print("\nĐang visualize lượng missing data...")

# Tính toán số lượng và phần trăm missing data
missing_data = df.isnull().sum()
missing_data = missing_data[missing_data > 0] # Chỉ lấy các cột có missing data
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_percent = missing_percent[missing_percent > 0]

missing_df = pd.DataFrame({
    'Missing Count': missing_data,
    'Missing Percent': missing_percent
}).sort_values(by='Missing Percent', ascending=False)

if not missing_df.empty:
    print("\nBảng tóm tắt Missing Data:")
    print(missing_df)

    # Cấu hình để hiển thị tiếng Việt nếu cần (đảm bảo font hỗ trợ)
    # plt.rcParams['font.family'] = 'DejaVu Sans' # Hoặc 'Arial Unicode MS', 'Liberation Sans' v.v.

    plt.figure(figsize=(12, 7))
    sns.barplot(x=missing_df.index, y='Missing Percent', data=missing_df, palette='viridis')
    plt.title('Phần trăm Missing Data theo Từng Biến', fontsize=16)
    plt.xlabel('Biến', fontsize=12)
    plt.ylabel('Phần trăm Missing (%)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 7))
    sns.barplot(x=missing_df.index, y='Missing Count', data=missing_df, palette='magma')
    plt.title('Số lượng Missing Data theo Từng Biến', fontsize=16)
    plt.xlabel('Biến', fontsize=12)
    plt.ylabel('Số lượng Missing', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
else:
    print("\nKhông có missing data nào được tìm thấy trong DataFrame.")


# =======================================================
# 8. Tính phần trăm số quan sát có ít nhất 1 missing data
# =======================================================
print("\nĐang tính phần trăm số quan sát có ít nhất 1 missing data...")
rows_with_missing = df.isnull().any(axis=1).sum()
total_rows = len(df)
percent_rows_with_missing = (rows_with_missing / total_rows) * 100

print(f"\nTổng số quan sát: {total_rows}")
print(f"Số quan sát có ít nhất 1 missing data: {rows_with_missing}")
print(f"Phần trăm quan sát có ít nhất 1 missing data: {percent_rows_with_missing:.2f}%")

print("\n--- Hoàn tất phân tích missing data ---")

# Hiển thị thông tin tổng quan cuối cùng
print("\nThông tin DataFrame sau tất cả các bước xử lý:")
print(df.info())
```
