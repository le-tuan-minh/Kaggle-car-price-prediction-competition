```python
import pandas as pd
```

```python
df = pd.read_excel("carprice_clean.xlsx", sheet_name=0, engine="openpyxl")
```

```python
df.info()
```

```python
df.drop(columns=['mileage', 'torque', 'mileage_numeric', 'mileage_unit', 'torque_unit_extracted'], inplace=True)
```

```python
df.info()
```

```python
# =======================================================
# Thư viện cần thiết
# =======================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# =======================================================
# 1. Phân loại biến
# =======================================================
target_variable = 'selling_price'
numerical_cols = [
    'year', 'km_driven', 'engine', 'max_power', 'seats',
    'mileage_kmpl_equivalent', 'torque_Nm'
]
categorical_cols = [
    'name', 'fuel', 'seller_type', 'transmission', 'owner'
]

if target_variable in numerical_cols:
    numerical_cols.remove(target_variable)

print(f"\nBiến mục tiêu: {target_variable}")
print(f"Các biến số (Numerical): {numerical_cols}")
print(f"Các biến phân loại (Categorical): {categorical_cols}")

# =======================================================
# 2. Phân tích Univariate
# =======================================================
print("\n--- Thực hiện Phân tích Univariate ---")

# 2.1. Biến số (Histogram và Box Plot)
for col in numerical_cols:
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Phân bố của {col}')
    plt.xlabel(col)
    plt.ylabel('Tần số')

    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[col], color='lightcoral')
    plt.title(f'Box Plot của {col}')
    plt.ylabel(col)

    plt.tight_layout()
    plt.show()

# 2.2. Biến phân loại (Count Plot)
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    if col == 'name':
        top_n = 20
        top_names = df[col].value_counts().nlargest(top_n).index
        sns.countplot(y=df[col], order=top_names, palette='viridis')
        plt.title(f'Số lượng xe theo {col} (Top {top_n})')
    else:
        sns.countplot(y=df[col], order=df[col].value_counts().index, palette='viridis')
        plt.title(f'Số lượng xe theo {col}')
    plt.xlabel('Số lượng')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()

# =======================================================
# 3. Phân tích Bivariate với selling_price
# =======================================================
print(f"\n--- Phân tích Bivariate với biến mục tiêu '{target_variable}' ---")

# 3.1. Numerical vs. selling_price
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=df[col], y=df[target_variable], alpha=0.6, color='darkblue')
    plt.title(f'{target_variable} vs. {col}')
    plt.xlabel(col)
    plt.ylabel(target_variable)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

# 3.2. Categorical vs. selling_price
for col in categorical_cols:
    plt.figure(figsize=(12, 7))
    if col == 'name':
        top_n = 20
        top_names = df[col].value_counts().nlargest(top_n).index
        df_filtered = df[df[col].isin(top_names)]
        sns.boxplot(x=df_filtered[col], y=df_filtered[target_variable], palette='coolwarm')
        plt.title(f'Phân bố {target_variable} theo {col} (Top {top_n})')
        plt.xticks(rotation=45, ha='right')
    else:
        sns.boxplot(x=df[col], y=df[target_variable], palette='coolwarm')
        plt.title(f'Phân bố {target_variable} theo {col}')
        plt.xticks(rotation=45, ha='right')
    plt.xlabel(col)
    plt.ylabel(target_variable)
    plt.tight_layout()
    plt.show()

# =======================================================
# 4. Heatmap Ma trận Tương quan
# =======================================================
print("\n--- Biểu đồ Tương quan giữa các biến số ---")
corr_matrix = df[numerical_cols + [target_variable]].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Ma trận Tương quan')
plt.tight_layout()
plt.show()

# =======================================================
# 5. Trực quan hóa t-SNE
# =======================================================
print("\n--- Trực quan hóa dữ liệu bằng t-SNE ---")

# Chuẩn bị dữ liệu cho t-SNE
selected_cols = numerical_cols + categorical_cols
df_tsne = df[selected_cols].dropna()

# Xử lý encoding & scaling
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), categorical_cols)
])

X_tsne = preprocessor.fit_transform(df_tsne)

# Giảm chiều với t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30, learning_rate=200)
X_embedded = tsne.fit_transform(X_tsne)

# Vẽ scatterplot
plt.figure(figsize=(10, 7))
sns.scatterplot(x=X_embedded[:, 0], y=X_embedded[:, 1], hue=df_tsne['fuel'], palette='tab10', alpha=0.7)
plt.title('t-SNE biểu diễn dữ liệu (màu theo fuel)')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.legend(title='Fuel')
plt.tight_layout()
plt.show()
```
